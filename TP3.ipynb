## Step 2 – Multi-Agent System Definition
Agents:
- DataAgent: loads images and metadata
- ImageAgent: preprocesses images
- MLAgent: trains models
- GameAgent: optimizes hyperparameters
- EvaluationAgent: computes metrics

```python
class DataAgent:
    def load_metadata(self, csv_path):
        return pd.read_csv(csv_path)

    def load_image_paths(self, img_root):
        images = []
        labels = []
        for label in ["fake", "real"]:
            folder = os.path.join(img_root, label)
            if not os.path.exists(folder):
                continue
            for f in os.listdir(folder):
                images.append(os.path.join(folder, f))
                labels.append(1 if label == "fake" else 0)
        return images, labels
```
